```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

personalidades = pd.read_csv("personalidades-nacionais.csv")

personalidades
```

```python
persona = personalidades.iloc[0:10,np.r_ [1,5:10]]

IPD = {'Perfil Avaliado': persona['PERFIL'], 
     'Indice de Popularidade Digital': persona['IPD'], 
     'Fama' : persona['FAMA'],
     'Engajamento' : persona['ENGAJ'],
     'Valencia' : persona['VALENCIA']
    }

IPD = pd.DataFrame(data=IPD)


ax = IPD.plot.bar(x = 'Perfil Avaliado', y = 'Indice de Popularidade Digital', color='#E8FC03',rot = 40, figsize=(15,10))
ax.xaxis.set_label_text("")
ax.yaxis.grid(color='gray', linestyle='dashed')
plt.tight_layout()
ax.set_title('Indice de Popularidade Digital das Top 10 Personalidades Brasileiras em 2021 ')
```

```python
FEV = persona.sort_index(axis=0 ,ascending=False)

fields = ['FAMA','ENGAJ','VALENCIA']
colors = ['#1D2F6F', '#8390FA', '#6EAF46']
labels = ['Fama', 'Engajamento', 'Valencia']

fig, ax = plt.subplots(1, figsize=(12, 10))

# plot bars
left = len(FEV) * [0]
for idx, name in enumerate(fields):
    plt.barh(FEV.PERFIL, FEV[name], left = left, color=colors[idx])
    left = left + FEV[name]

# title, legend, labels
plt.title('Top 10 personalidades brasileiras no ano de 2021\n', loc='left')
plt.legend(labels, bbox_to_anchor=([0.55, 1, 0, 0]), ncol=4, frameon=False)

plt.ylim(-0.5, ax.get_yticks()[-1] + 0.5)
ax.xaxis.grid(color='gray', linestyle='dashed')

plt.show()
```
